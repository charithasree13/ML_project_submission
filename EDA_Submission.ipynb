## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# ==========================================
# BASIC DATA CHECK
# ==========================================
try:
    print("\nDataset Shape:", data.shape)
    print("\nColumns:", data.columns)
    print("\nMissing Values:\n", data.isnull().sum())
except Exception as e:
    print("Error in data inspection:", e)


# ==========================================
# HANDLE MISSING VALUES
# ==========================================
try:
    # Fill missing values (if any)
    # It seems 'Incident_Counts' is entirely NaN. We might need to decide how to fill this,
    # or if it's meant to be calculated later. For now, using ffill might not be appropriate
    # if it's the only NaN column. Let's adjust this for the specific column or drop it if it's not used.
    # Given 'Incident_Counts' is entirely NaN, ffill will not work. Let's keep it as is for now if it's to be computed.
    # If other columns had NaNs, ffill would be applicable. Since only Incident_Counts has NaNs and it's all of them,
    # ffill won't change anything for that column.
    if data['Incident_Counts'].isnull().all():
        print("Incident_Counts is entirely missing, not filling with ffill.")
    else:
        data.fillna(method='ffill', inplace=True)
        print("Missing values handled ")
except Exception as e:
    print("Error handling missing values:", e)


# ==========================================
# REMOVE DUPLICATES
# ==========================================
try:
    before = data.shape[0]
    data.drop_duplicates(inplace=True)
    after = data.shape[0]
    print(f"Duplicates removed: {before - after} rows")
except Exception as e:
    print("Error removing duplicates:", e)


# ==========================================
# DATA TYPE CONVERSION & FEATURE ENGINEERING (Combined)
# ==========================================
try:
    # Create a 'Date' column from 'YEAR' and 'MONTH' since 'Date' column is missing
    data['Date'] = pd.to_datetime(data['YEAR'].astype(str) + '-' + data['MONTH'].astype(str) + '-01')
    print("Date column created and converted ")

    data['Year'] = data['Date'].dt.year
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day
    data['DayOfWeek'] = data['Date'].dt.day_name()
    print("New features created ")
except Exception as e:
    print("Error in creating Date or features:", e)


# ==========================================
# SORT DATA
# ==========================================
try:
    data = data.sort_values(by='Date')
    print("Data sorted by Date ")
except Exception as e:
    print("Error sorting data:", e)


# ==========================================
# CREATE AGGREGATED DATA (TIME SERIES)
# ==========================================
try:
    # Since Incident_Counts is all NaN, let's assume we want to count incidents by TYPE, YEAR, MONTH for now
    # and then aggregate daily. The original daily_crime creation was just counting rows per date.
    # Let's recreate daily_crime as intended by the original logic: counting all incidents by date.
    daily_crime = data.groupby('Date').size().reset_index(name='Crime_Count')
    print("Aggregated data created ")
except Exception as e:
    print("Error creating aggregated data:", e)


# ==========================================e
# OUTLIER DETECTION (IQR METHOD)
# ==========================================
try:
    Q1 = daily_crime['Crime_Count'].quantile(0.25)
    Q3 = daily_crime['Crime_Count'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = daily_crime[
        (daily_crime['Crime_Count'] < lower_bound) |
        (daily_crime['Crime_Count'] > upper_bound)
    ]

    print(f"Outliers detected: {len(outliers)}")
except Exception as e:
    print("Error in outlier detection:", e)


# ==========================================
# FINAL DATA CHECK
# ==========================================
try:
    print("\nFinal Dataset Preview:")
    print(data.head())

    print("\nDaily Crime Data Preview:")
    print(daily_crime.head())
except Exception as e:
    print("Error displaying data:", e)


# ==========================================
# COMPLETION MESSAGE
# ==========================================
print("\nData Wrangling Completed Successfully ")
```
